```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'student_management.csv'
data = pd.read_csv(file_path)

pd.set_option('display.max_rows', None)  # Show kar raha all rows ko
pd.set_option('display.max_columns', None) # coulmn ko
pd.set_option('display.width', None)  
pd.set_option('display.colheader_justify', 'center')  
```

```python

data.head(5)
```

```python
data.shape
```

```python
def clean_data(df):
    df.rename(columns={"Nacionality": "Nationality"}, inplace=True)
    df.replace({
        'Gender': {0: 'Male', 1: 'Female'},
        'Marital status': {0: 'Single', 1: 'Married', 2: 'Divorced'},
        'Application mode': {0: 'Offline', 1: 'Online', 17: 'Hybrid', 39: 'In-Person', 15: "Virtual", 10: "Physical"},
        'Daytime/evening attendance\t': {0: 'Evening', 1: 'Daytime'},
        'Nationality': {1: 'Pakistani', 2: 'Turkish', 105: "India", 22: "Saudi"},
        "Mother's qualification": {1: 'Graduate', 2: 'Undergraduate', 19: "Graduate", 37: "Graduate", 38: "Graduate"},
        "Father's qualification": {1: 'Graduate', 2: 'Undergraduate', 19: "Graduate", 37: "Graduate", 38: "Graduate", 12: "Undergraduate", 3: "Undergraduate"},
        "Mother's occupation": {1: 'Teacher', 2: 'Engineer', 5: "Doctor", 3: "Lawyer", 9: "Home wife", 7: "Businessman"},
        "Father's occupation": {1: 'Doctor', 2: 'Lawyer', 3: "Lawyer", 9: "Businessman", 4: "Navy"},
        'International': {0: 'No', 1: 'Yes'}
    }, inplace=True)
    return df

data = clean_data(data)


data.head(5)
```

```python
data.tail(5)
```

```python
total_students = len(data)
print("Total Student:",total_students)

plt.bar(['Total Students'], [total_students], color='grey')

 
```

```python
student_status = data['Target'].value_counts()
print(student_status)

```

```python
chart = data['Target'].value_counts().plot(kind="pie",color="red")
```

```python
average_grade = data['Curricular units 2nd sem (grade)'].mean()
print("Average of all students:", average_grade)
```

```python
# Grades 
high_grade_students = data[data['Curricular units 2nd sem (grade)'] > 15]
print(data['Curricular units 2nd sem (grade)'])
```

```python
counts = data.groupby(['Gender', 'Scholarship holder']).size().unstack()
counts.plot(kind='bar', stacked=True, color=['lightgreen', 'purple'])
plt.title('Gender vs Scholarship Holder')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python

counts = data.groupby(['Marital status', 'Application mode']).size().unstack()
counts.plot(kind='bar', stacked=True, color=['lightblue', 'orange'])
plt.title('Marital Status vs Application Mode')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
counts = data.groupby(['Educational special needs', 'Displaced']).size().unstack()
counts.plot(kind='bar', stacked=True, color=['lightgreen', 'red'])
plt.title('Educational Special Needs vs Displaced')
plt.xlabel('Educational Special Needs')
plt.ylabel('Count')
plt.show()
```
