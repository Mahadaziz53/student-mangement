```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'student_management.csv'
data = pd.read_csv(file_path)

pd.set_option('display.max_rows', None)  # Show kar raha all rows ko
pd.set_option('display.max_columns', None) # columns ko show karwa rahi
pd.set_option('display.width', None)  
pd.set_option('display.colheader_justify', 'center')  

```

```python

data.head(5)
```

```python
data.shape
```

```python
data.rename(columns={"Nacionality": "Nationality"}, inplace=True)
data.rename(columns={"GDP":"CGPA"}, inplace=True)
data['Gender'] = data['Gender'].replace({0: 'Male', 1: 'Female'})
data['Marital status'] = data['Marital status'].replace({0: 'Single', 1: 'Married'})
data['Marital status'] = data['Marital status'].replace({2: 'Divorced',})
data['Application mode'] = data['Application mode'].replace({0: 'Offline',1: 'Online',17: 'Hybrid',  
    39: 'In-Person',15:"Virtual",10:"Physical"})
data['Daytime/evening attendance\t'] = data['Daytime/evening attendance\t'].replace({0: 'Evening', 1: 'Daytime'})


data['Nationality'] = data['Nationality'].replace({1: 'Pakistani', 2: 'Turkish',105:"India",22:"Saudi"})
data["Mother's qualification"] = data["Mother's qualification"].replace({1: 'Graduate', 2: 'Undergraduate',19:"Graduate",37:"Graduate",38:"Graduate"})
data["Father's qualification"] = data["Father's qualification"].replace({1: 'Graduate', 2: 'Undergraduate',19:"Graduate",37:"Graduate",38:"Graduate",12:"Undergraduate",3:"Undergraduate"})
data["Mother's occupation"] = data["Mother's occupation"].replace({1: 'Teacher', 2: 'Engineer',5:"Docter",3:"Lawyer",9:"Home wife",7:"Bussinessman"})
data["Father's occupation"] = data["Father's occupation"].replace({1: 'Doctor', 2: 'Lawyer',3:"Lawyer",9:"Bussinessman",4:"Navy"})
data['International'] = data['International'].replace({0: 'No', 1: 'Yes'})

data.head(5)
```

```python
data.tail(5)
```

```python
total_students = len(data)
print("Total Student:",total_students)

plt.bar(['Total Students'], [total_students], color='grey')


```
